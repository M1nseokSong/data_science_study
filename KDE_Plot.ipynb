```python
!pip install seaborn==0.9.0
import pandas as pd
import seaborn as sns

body_df = pd.read_csv('body.csv', index_col=0)
body_df.head()
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

```python
sns.kdeplot(body_df['Height'], bw=0.5) #bw는 bandwidth
```

```python
sns.distplot(body_df['Height'], bins=15)
```

```python
sns.violinplot(y=body_df['Height'])
```

```python
sns.kdeplot(body_df['Height'], body_df['Weight']) #등고선 모양
```
