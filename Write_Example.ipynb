1. ID 1의 무게를 200으로 변경하세요.
2. ID 21의 row를 삭제하세요.
3. ID 20의 row를 추가하세요. ID 20의 키는 70, 무게는 200입니다. **굵은 텍스트**


```python
import pandas as pd

df = pd.read_csv('body_imperial1.csv', index_col=0)

df.loc[1, 'Weight (Pound)'] = 200
df.drop(21, axis='index', inplace=True)
df.loc[20] = [70, 200]

df
```
